Se utilizará solo algunas variables de ambas encuestas.  
- Para ENDES 2022, las variables ```edadr sexo dep zona1  P04 P05 P06```. 
- Para ENDES 2024, las variables ``` edadr sexo dep zona1  P07 P08 P09 P10 ```.

 
 PARA ENDES 2022:
- edadr: grupos de edad
- sexo: sexo
- dep: departamento
- zona2: subzona de residencia
- P04: Percepcepcion sobre el nivel de desigualdad económica en el Perú. 
- P05: Percepción sobre la diferencia entre ricos y pobres en el Perú.
- P06: Percepción sobre su status social.

 PARA ENDES 2024:
- edadr: grupos de edad
- sexo: sexo
- dep: departamento
- zona2: subzona de residencia
- P07: Percepcepcion sobre el nivel de desigualdad económica en el Perú. 
- P08: Percepción sobre la diferencia entre ricos y pobres en el Perú.
- P09: Percepición sobre si el lugar de vivienda en la infancia otorga mayor oportunidad.
- P10: Factores que ayudan a tener un país más igualitario. 


````python
dfenades22 = df22.filter(items=['edadr', 'sexo', 'dep', 'zona2',  'p04', 'p05', 'p06'])
dfenades22.columns
````

````python
dfenades24 = df24.filter(items=['edadr', 'sexo', 'dep', 'zona2',  'P07', 'P08', 'P09', 'P10'])
dfenades24.columns
````

````python
dfenades22
````

NOTA: Limpiaremos las variables donde el valor 91 o 99 significa NS/NP.

````python
#PARA ENDES 2022
end22_filtrado = dfenades22[
    (dfenades22["p04"].isin([1, 2, 3, 4])) &
    (dfenades22["p05"].isin([1, 2, 3])) &
    (dfenades22["p06"].isin([1, 2, 3, 4])) &
    (dfenades22["sexo"].isin([1, 2]))
]
end22_filtrado
````

````python
#PARA ENDES 2024
end24_filtrado = dfenades24[
    (dfenades24["P07"].isin([1, 2, 3, 4])) &
    (dfenades24["P08"].isin([1, 2, 3])) &
    (dfenades24["P09"].isin([1, 2, 3, 4])) &
    (dfenades24["P10"].isin([1, 2, 3, 4, 5])) &
    (dfenades24["sexo"].isin([1, 2]))
]
end24_filtrado
````

PARA LA ENDES 2022, CREAREMOS 4 VARIABLES QUE CONTENGAN LAS ETIQUETAS DE:  SEXO, P04, P05 Y P06. Esto debido a que están codificadas como floats. 

````python

#PARA SEXO
end22_filtrado.loc[:, "SEXO"] = end22_filtrado["sexo"].replace({
    1: "Hombre",
    2: "Mujer"
})
#PARA DEPARTAMENTO

end22_filtrado.loc[:, "DEPARTAMENTO"] = end22_filtrado["dep"].replace({
    1: "Amazonas",
    2: "Áncash",
    3: "Apurímac",
    4: "Arequipa",
    5: "Ayacucho",
    6: "Cajamarca",
    8: "Cusco",
    9: "Huancavelica",
    10: "Huánuco",
    11: "Ica",
    12: "Junín",
    13: "La Libertad",
    14: "Lambayeque",
    15: "Lima",
    16: "Loreto",
    17: "Madre de Dios",
    18: "Moquegua",
    19: "Pasco",
    20: "Piura",
    21: "Puno",
    22: "San Martín",
    23: "Tacna",
    24: "Tumbes",
    25: "Ucayali"
})


#PARA P04
end22_filtrado.loc[:, "Nivel_desigualdad_peru"] = end22_filtrado["p04"].replace({
    1: "Mucho",
    2: "Algo",
    3: "Poco",
    4: 'Nada'
})

#PARA P05
end22_filtrado.loc[:, "Dife_entre_ricos_pobres"] = end22_filtrado["p05"].replace({
    1: "Ha aumentado",
    2: "Se mantiene igual",
    3: "Ha disminuido"
})

#PARA P06

end22_filtrado.loc[:,"Percepcion_status_social"] = end22_filtrado["p06"].replace({
    1: "Alta",
    2: "Media alta",
    3: "Media Baja",
    4: "Baja"
})
````

````python
end22_filtrado
````

PARA LA ENDES 2024, CREAREMOS 4 VARIABLES QUE CONTENGAN LAS ETIQUETAS DE SEXO, P07, P08 Y P09. Esto debido a que están codificadas como floats.

````python
#PARA SEXO
end24_filtrado.loc[:, "SEXO"] = end24_filtrado["sexo"].replace({
    1: "Hombre",
    2: "Mujer"
})

#PARA DEPARTAMENTO

end24_filtrado.loc[:, "DEPARTAMENTO"] = end24_filtrado["dep"].replace({
    1: "Amazonas",
    2: "Áncash",
    3: "Apurímac",
    4: "Arequipa",
    5: "Ayacucho",
    6: "Cajamarca",
    8: "Cusco",
    9: "Huancavelica",
    10: "Huánuco",
    11: "Ica",
    12: "Junín",
    13: "La Libertad",
    14: "Lambayeque",
    15: "Lima",
    16: "Loreto",
    17: "Madre de Dios",
    18: "Moquegua",
    19: "Pasco",
    20: "Piura",
    21: "Puno",
    22: "San Martín",
    23: "Tacna",
    24: "Tumbes",
    25: "Ucayali"
})

#PARA P07
end24_filtrado.loc[:, "Nivel_desigualdad_peru2"] = end24_filtrado["P07"].replace({
    1: "Nada",
    2: "Poco",
    3: "Algo",
    4: 'Mucho'
})

#PARA P08
end24_filtrado.loc[:, "Dife_entre_ricos_pobres2"] = end24_filtrado["P08"].replace({
    1: "Ha aumentado",
    2: "Se mantiene igual",
    3: "Ha disminuido"
})

#PARA P09

end24_filtrado.loc[:,"Percepcion_lugar_infancia"] = end24_filtrado["P09"].replace({
    1: "Nada importante",
    2: "Poco importante",
    3: "Algo importante",
    4: "Muy importante"
})
````

````python
end24_filtrado
````

Ahora haremos una tabulacion cruzada, donde veremos el número de personas, por percepción sobre el nivel de desigualdad económica en el Perú, que existe en cada departamento.

````python
#PARA ENDES 2022
tabla = pd.pivot_table(end22_filtrado, 
               index = 'DEPARTAMENTO', # filas
                  columns = 'Nivel_desigualdad_peru', # columnas
                     aggfunc='size'
                           )  

tabla
````

````python
#PARA ENDES 2024
tabla2 = pd.pivot_table(end24_filtrado, 
               index = 'DEPARTAMENTO', # filas
                  columns = 'Nivel_desigualdad_peru2', # columnas
                     aggfunc='size'
                           )  

tabla2
````

#### GRAFICAMOS:

````python
chart = alt.Chart(end22_filtrado).mark_bar().encode(
    x=alt.X("DEPARTAMENTO:N", title="Departamento"),  
    y=alt.Y("count()", title="Número de respuestas"),  
    color=alt.Color("Nivel_desigualdad_peru", title="Nivel de desigualdad económica",
                    scale=alt.Scale(scheme="paired"),
                    sort="ascending")).properties(
    title="Distribución de la percepción sobre la desigualdad económica en el Perú por Departamento - ENDES 2022")
chart
````
